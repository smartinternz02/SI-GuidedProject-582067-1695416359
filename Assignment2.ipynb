```python
# IMPORTNG LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# LOADING DATASET
df = pd.read_csv('House Price India.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# UNIVARIATE ANALYSIS 1
# for categorical data
cd = ['number of bedrooms','number of floors','waterfront present','number of views',
      'condition of the house','grade of the house','Number of schools nearby']
for i in cd:
  plt.figure(figsize= (7,7))
  plt.title(i)
  sns.countplot(data = df, x = i)
  plt.show()
```

```python
# UNIVARIATE ANALYSIS 2
# for numeric data
l=['living area','lot area','Area of the house(excluding basement)','Area of the basement']
for i in l:
  plt.title(i)
  plt.hist(df['living area'], bins = 50)
  plt.show()
```

```python
# UNIVARIATE ANALYSIS 3
# for price
sns.displot(df['Price'])
plt.title('Price Distribution')
plt.show()
```

```python
# BIVARIATE ANALYSIS 1
x_cols = ['number of bedrooms', 'number of floors', 'condition of the house','number of views']
y_cols = ['number of bathrooms','grade of the house']
for i in x_cols:
  for j in y_cols:
    sns.barplot(data=df, x=i, y=j)
    plt.title(i+' vs '+j)
    plt.show()
```

```python
# BIVARIATE ANALYSIS 2
x = ['number of bedrooms','living area','Area of the basement','Built Year']
for i in x:
  plt.title("Scatterplot of "+i+" and Price", fontsize=15)
  plt.xlabel(i, fontsize=15)
  plt.ylabel("Price", fontsize=15)
  plt.scatter(x= df[i], y= df.Price)
  plt.show()
```

```python
# MULTIVARIATE ANALYSIS 1
plt.figure(figsize=(12,12))
plt.title('CORRELATION BETWEEN DIFFERENT FEATURES')
sns.heatmap(df.corr(), square= True, vmax=0.8)
```

```python
# MULTIVARIATE ANALYSIS 2
columns = ['living area','lot area','Area of the house(excluding basement)','Area of the basement',
           'Built Year','Lattitude','Longitude','living_area_renov','lot_area_renov','Distance from the airport','Price']
sns.pairplot(df[columns],size=2,kind="scatter")
plt.show()
```
